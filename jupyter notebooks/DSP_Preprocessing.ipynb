```python
#@title Lemmatization (English) & Stemming (Malay)
from nltk.corpus import wordnet

#part-of-speech tag
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    else:
        return None
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

english_lemmatizer = WordNetLemmatizer()

def english_clean_text(text):
    tokens = nltk.word_tokenize(text)
    tagged = nltk.pos_tag(tokens)

    full_text = ''

    for word, tag in tagged:
        wordntag = get_wordnet_pos(tag)
        if wordntag is None:
            lemmat = english_lemmatizer.lemmatize(word)
        else:
            lemmat = english_lemmatizer.lemmatize(word, pos=wordntag)
        full_text += lemmat + ' '
    return full_text
    
english_round2 = lambda x: english_clean_text(x)
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
dfEng.Tweets = dfEng.Tweets.apply(english_round2)
```
